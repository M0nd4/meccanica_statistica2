## 1. Single Flip

**Metropolis importance sampling Monte Carlo**

(1) Chose an initial state

(2) Choose a site $i$

(3) Caclulate the energy change $\Delta E$ which results if the spin at site $i$ is overturned

(4) Generate a random number $r$ such that $0 < r < 1$

(5) if $r < \exp[-\beta\Delta E]$, flip the spin

(6) Go to the next site and go to (3)

```python
def plot(T = 1,L = 128, fn = 10):
    
    def x_y(k, L):
        y = k // L
        x = k - y * L
        return x, y
    
    N = L * L
    
    # Neighbours of a specific index
    nbr = {i : ((i // L) * L + (i + 1) % L, (i + L) % N,
                (i // L) * L + (i - 1) % L, (i - L) % N) \
                                        for i in range(N)}

    # Load From File. We need this because effective Sampling needs lof of steps
    filename = 'local_'+ str(L) + '_' + str(T) + "_" + '.txt'
    if os.path.isfile(filename):
        f = open(filename, 'r')
        S = []
        for line in f:
            S.append(int(line))
        f.close()
        print( 'starting from file', filename)
    else:
        S = [random.choice([1, -1]) for k in range(N)]
        print( 'starting from scratch')
        
        
    nsteps = fn
    
    # We choose k = 1 and J = 1 => beta = 1 / T
    beta = 1.0 / T
    
    for step in range(nsteps):
        
        # Choosing a random site i
        k = random.randint(0, N - 1)
        
        # Calculating the energy difference if we would flip the spin of the site i
        delta_E = 2.0 * S[k] * sum(S[nn] for nn in nbr[k])
        
        # Applying the Transition probability
        if random.uniform(0.0, 1.0) < math.exp(-beta * delta_E):
            
            #Flip the spin
            S[k] *= -1
            
        # Printing the progress
        if step % (nsteps/5) == 0 :
            print("step = %d / %d \t %.3f%%"%(step,nsteps,100 * step/nsteps))
    
    # Plotting
    conf = [[0 for x in range(L)] for y in range(L)]
    for k in range(N):
        x, y = x_y(k, L)
        conf[x][y] = S[k]

    plt.figure(figsize = (15,15))
    plt.imshow(conf, extent = [0, L, 0, L], interpolation='nearest')
    plt.set_cmap('hot')
    plt.title('Configuration at T = %.3f and L=%d'%(T,L))
    plt.show()
    
    # Saving
    f = open(filename, 'w')
    for a in S:
       f.write(str(a) + '\n')
    f.close()
```

## Energy Distribution

```python
def plot_energies(T = 1.0, L = 16, fn = 1000):
    
    def x_y(k, L):
        y = k // L
        x = k - y * L
        return x, y

    N = L * L

    # Neighbours of a specific index
    nbr = {i : ((i // L) * L + (i + 1) % L, (i + L) % N,
                (i // L) * L + (i - 1) % L, (i - L) % N) \
                                        for i in range(N)}


    S = [random.choice([1, -1]) for k in range(N)]
    
    def energy(S):
        E = 0.0
        for k in range(N):
            E -=  S[k] * sum(S[nn] for nn in nbr[k])
        return 0.5 * E


    # Initial Energy
    E0 = energy(S)
                
    nsteps = fn


    # We choose k = 1 and J = 1 => beta = 1 / T
    beta = 1.0 / T
    E = np.zeros(nsteps)+ E0
    print(" ----------- ")
    print("new run at T = %.3f and L = %d"%(T,L))
    for step in range(nsteps):
        
        # Choosing a random site i
        k = random.randint(0, N - 1)
        
        # Calculating the energy difference if we would flip the spin of the site i
        delta_E = 2.0 * S[k] * sum(S[nn] for nn in nbr[k])
        
        # Applying the Transition probability
        if random.uniform(0.0, 1.0) < math.exp(-beta * delta_E):
            
            #Flip the spin
            S[k] *= -1
            E[step] += delta_E
            
        # Printing the progress
        if step % (nsteps/5) == 0 :
            print("step = %d / %d \t %.3f%%"%(step,nsteps,100 * step/nsteps))
    print("mean energy is ",np.mean(E))        
    plt.figure()
    plt.title("Histogram of the Energies at T = %.3f and L = %d"%(T,L))
    plt.hist(E)
    plt.show()
```

# 2. Cluster Algorithms

**Swendsen-Wang algorithm for the Ising Model**

1. Randomly choose a site 
2. Chose all nearest neighbours with the same spin which are not already in the cluster, with probability $p = 1 - \exp({-J \beta})$
3. Repeat second step, using all the neighbours which were chosen, until no neigbhours are found anymore
4. Find all Clusters
5. Flip each Cluster with probability $q$
5. Go to 1 for next iteration

**Wolff cluster flipping method for the Ising model**

1. Randomly choose a site 
2. Chose all nearest neighbours with the same spin which are not already in the cluster, with probability $p = 1 - \exp({-J \beta})$
3. Repeat second step, using all the neighbours which were chosen, until no neigbhours are found anymore
4. Flip all spins in the cluster
5. Go to 1 for next iteration

```python
def wolff_algorithm(T = 2.0, L = 6, nsteps = 10000, S = 0):
    
    N = L * L
    # Initial Configuration
    if S == 0:
        S = [random.choice([1, -1]) for k in range(N)]
        
    # Function for the Energy
    def energy(S, N, nbr):
        E = 0.0
        for k in range(N):
            E -=  S[k] * sum(S[nn] for nn in nbr[k])
        return 0.5 * E


    
    # Defining Nearest neighbours
    nbr = {i : ((i // L) * L + (i + 1) % L, (i + L) % N,
                (i // L) * L + (i - 1) % L, (i - L) % N)
                                        for i in range(N)}
    
    # Probability of chosing a neighbour
    p  = 1.0 - math.exp(-2.0 / T)

    # Initial Energy
    E = [energy(S, N, nbr)]
    
    
    for step in range(nsteps):
        # Choosing a random site
        k = random.randint(0, N - 1)
        
        # Initializing the cluster and the pocket
        Pocket, Cluster = [k], [k]
        
        # We look for neighbours as long as possible
        while Pocket != []:
            # We choose a site from our pocket
            j = random.choice(Pocket)
            
            # Checking all the nearest neighbours of the site
            for l in nbr[j]:
                # Only nearest neighbours with same spin are considered
                if S[l] == S[j] and l not in Cluster \
                       and random.uniform(0.0, 1.0) < p:
                    # Adding to the our pocket, since we want the new nearest neighbours as well
                    Pocket.append(l)
                    Cluster.append(l)
            Pocket.remove(j)
        # Flipping the whole cluster
        for j in Cluster:
            S[j] *= -1
        E.append(energy(S, N, nbr))
    return (E,S)

def plot_conf(L, S):
    N = L * L
    # Plotting
    conf = [[0 for x in range(L)] for y in range(L)]
    for k in range(N):
        x, y = x_y(k, L)
        conf[x][y] = S[k]

    plt.figure(figsize = (15,15))
    plt.imshow(conf, extent = [0, L, 0, L], interpolation='nearest')
    plt.set_cmap('hot')
    plt.title('Configuration at T = %.3f and L=%d'%(T,L))
    plt.show()
    
```

# Magnetization at different Temperature Regimes

```python
def wolff_algorithm2(T = 2.0, L = 6, nsteps = 10000, S = 0):
    N = L * L
    # Initial Configuration
    if S == 0:
        S = [random.choice([1, -1]) for k in range(N)]
        
    # Function for the Energy
    def energy(S, N, nbr):
        E = 0.0
        for k in range(N):
            E -=  S[k] * sum(S[nn] for nn in nbr[k])
        return 0.5 * E

    
    # Defining Nearest neighbours
    nbr = {i : ((i // L) * L + (i + 1) % L, (i + L) % N,
                (i // L) * L + (i - 1) % L, (i - L) % N)
                                        for i in range(N)}
    
    # Probability of chosing a neighbour
    p  = 1.0 - math.exp(-2.0 / T)

    # Initial magnetization
    M = [np.mean(S)]
    
    
    for step in range(nsteps):
        # Choosing a random site
        k = random.randint(0, N - 1)
        
        # Initializing the cluster and the pocket
        Pocket, Cluster = [k], [k]
        
        # We look for neighbours as long as possible
        while Pocket != []:
            # We choose a site from our pocket
            j = random.choice(Pocket)
            
            # Checking all the nearest neighbours of the site
            for l in nbr[j]:
                # Only nearest neighbours with same spin are considered
                if S[l] == S[j] and l not in Cluster \
                       and random.uniform(0.0, 1.0) < p:
                    # Adding to the our pocket, since we want the new nearest neighbours as well
                    Pocket.append(l)
                    Cluster.append(l)
            Pocket.remove(j)
        # Flipping the whole cluster
        for j in Cluster:
            S[j] *= -1
        M.append(np.mean(S))
    return M
```

# Calculating the Correlation Time $\tau$ with the magnetization

## $\phi(t) = \frac{\langle M(t_0) M(t)\rangle - {\langle M \rangle}^2}{\langle M^2 \rangle - {\langle M \rangle}^2}$ 

## $\phi(t) \propto  \exp(-t/\tau)$

```python
def wolff_Correlation(T = 2.0, L = 6, nsteps = 10000, S = 0):
    N = L * L
    # Initial Configuration
    if S == 0:
        S = [random.choice([1, -1]) for k in range(N)]
        
    # Function for the Energy
    def energy(S, N, nbr):
        E = 0.0
        for k in range(N):
            E -=  S[k] * sum(S[nn] for nn in nbr[k])
        return 0.5 * E

    
    # Defining Nearest neighbours
    nbr = {i : ((i // L) * L + (i + 1) % L, (i + L) % N,
                (i // L) * L + (i - 1) % L, (i - L) % N)
                                        for i in range(N)}
    
    # Probability of chosing a neighbour
    p  = 1.0 - math.exp(-2.0 / T)

    # Initial magnetization
    
    S0 = np.array(S)
    
    corr = []
    
    
    for step in range(nsteps):
        # Choosing a random site
        k = random.randint(0, N - 1)
        
        # Initializing the cluster and the pocket
        Pocket, Cluster = [k], [k]
        
        # We look for neighbours as long as possible
        while Pocket != []:
            # We choose a site from our pocket
            j = random.choice(Pocket)
            
            # Checking all the nearest neighbours of the site
            for l in nbr[j]:
                # Only nearest neighbours with same spin are considered
                if S[l] == S[j] and l not in Cluster \
                       and random.uniform(0.0, 1.0) < p:
                    # Adding to the our pocket, since we want the new nearest neighbours as well
                    Pocket.append(l)
                    Cluster.append(l)
            Pocket.remove(j)
        # Flipping the whole cluster
        for j in Cluster:
            S[j] *= -1
            
        # Calculating the correlation function
        S_ = np.array(S)
        corr.append(np.mean(S0*S_))
    
    return corr
```

### Relaxationtime $\phi$ for different Lengths $L$

```python
for L in np.arange(12,60,8):
    corr = wolff_Correlation(T = 2.26, L=L, nsteps = 300)
    plt.figure()
    plt.title("L=%d"%L, fontsize = 20)
    plt.plot(corr,marker="x")
    plt.yscale("log")
    plt.ylabel("$\phi$",fontsize = 18)
    plt.xlabel("t",fontsize = 14)
    plt.show()
```

## Comparing  the Correlation time of the Single-Flip with the Cluster-Flip Algorithm

```python
threshold = 0.2
def wolff_tau(T = 2.0, L = 6, S = 0):
    N = L * L
    # Initial Configuration
    if S == 0:
        S = [random.choice([1, -1]) for k in range(N)]
        
    # Function for the Energy
    def energy(S, N, nbr):
        E = 0.0
        for k in range(N):
            E -=  S[k] * sum(S[nn] for nn in nbr[k])
        return 0.5 * E

    
    # Defining Nearest neighbours
    nbr = {i : ((i // L) * L + (i + 1) % L, (i + L) % N,
                (i // L) * L + (i - 1) % L, (i - L) % N)
                                        for i in range(N)}
    
    # Probability of chosing a neighbour
    p  = 1.0 - math.exp(-2.0 / T)

    # Initial magnetization
    
    S0 = np.array(S)
    
    corr = []
    done = False
    t = 0
    
    while done==False:
        t += 1
        # Choosing a random site
        k = random.randint(0, N - 1)
        
        # Initializing the cluster and the pocket
        Pocket, Cluster = [k], [k]
        
        # We look for neighbours as long as possible
        while Pocket != []:
            # We choose a site from our pocket
            j = random.choice(Pocket)
            
            # Checking all the nearest neighbours of the site
            for l in nbr[j]:
                # Only nearest neighbours with same spin are considered
                if S[l] == S[j] and l not in Cluster \
                       and random.uniform(0.0, 1.0) < p:
                    # Adding to the our pocket, since we want the new nearest neighbours as well
                    Pocket.append(l)
                    Cluster.append(l)
            Pocket.remove(j)
        # Flipping the whole cluster
        for j in Cluster:
            S[j] *= -1
            
        # Calculating the correlation function
        S_ = np.array(S)
        if np.mean(S0*S_) < threshold:
            return (-t/np.log(threshold))
        
def singleflip_tau(T = 1.0, L = 16):
    
    def x_y(k, L):
        y = k // L
        x = k - y * L
        return x, y

    N = L * L

    # Neighbours of a specific index
    nbr = {i : ((i // L) * L + (i + 1) % L, (i + L) % N,
                (i // L) * L + (i - 1) % L, (i - L) % N) \
                                        for i in range(N)}


    S = [random.choice([1, -1]) for k in range(N)]
    S0 = np.array(S)

    # We choose k = 1 and J = 1 => beta = 1 / T
    beta = 1.0 / T  
    
    t = 0
    done = False
    
    while done == False:
        t += 1
        
        # Choosing a random site i
        k = random.randint(0, N - 1)
        
        # Calculating the energy difference if we would flip the spin of the site i
        delta_E = 2.0 * S[k] * sum(S[nn] for nn in nbr[k])
        
        # Applying the Transition probability
        if random.uniform(0.0, 1.0) < math.exp(-beta * delta_E):       
            #Flip the spin
            S[k] *= -1
            S_ = np.array(S)
            if np.mean(S0*S_) < threshold:
                return (-t/np.log(threshold))
            
```

```python
tau_wolf = []
tau_single = []
T = 2.26
LL = np.arange(4,100,2)
for L in LL:
    print("Calculating L=%d"%L)
    tau_wolf   += [wolff_tau(T, L)]
    tau_single += [singleflip_tau(T,L)]
```

```python
def gerade(x,y):
    x=np.array(x)
    y=np.array(y)
    b=np.sum((x-np.mean(x))*y)/sum((x-np.mean(x))**2)
    a=np.mean(y)-b*np.mean(x)
    n=len(x)
    sy=np.sqrt(np.sum((y-b*x-a)**2)/(n-2))
    sa=sy*np.sqrt(1./n+np.mean(x)**2/np.sum((x-np.mean(x))**2))
    sb=sy*np.sqrt(1./np.sum((x-np.mean(x))**2))
    return (a,b,sa,sb,sy)

plt.figure(figsize = (12,12))

plt.title("Scaling of the Correlation Time",fontsize = 35)
plt.scatter(LL,tau_wolf,marker="x",c="r")
(a,b1,sa,sb,sy) = gerade(np.log(LL),np.log(tau_wolf))
plt.plot(LL, np.exp(a + b1*np.log(LL)),c="c")

plt.scatter(LL,tau_single,marker="x",c="g")
(a,b2,sa,sb,sy) = gerade(np.log(LL),np.log(tau_single))
plt.plot(LL, np.exp(a + b2*np.log(LL)), c= "b")
captions = ["Fit of Wolff with $z=%.3f$"%b1,"Fit of Singleflip with $z=%.3f$"%b2, "Data of Wolff","Data of Singleflip"]
plt.legend(captions,2,fontsize = 20)
plt.yscale("log")
plt.xscale("log")
plt.xlim(1,max(LL)*1.1)
plt.ylabel("$\\tau$",fontsize = 30)
plt.xlabel("Length $L$",fontsize = 18)
plt.grid(True,which="both",ls="-")
plt.savefig("corrtime.png")
plt.show()
```

# Calculation of the Specific Heat $C_v$, Skaling Laws

```python
def energy(S, N, nbr):
    E = 0.0
    for k in range(N):
        E -=  S[k] * sum(S[nn] for nn in nbr[k])
    return 0.5 * E


def x_y(k, L):
    y = k // L
    x = k - y * L
    return x, y
CV = []
for L in [2,4,8,16,32]:
    T = 2.27
    N = L * L
    nbr = {i : ((i // L) * L + (i + 1) % L, (i + L) % N,
                (i // L) * L + (i - 1) % L, (i - L) % N) \
                                        for i in range(N)}
    filename = 'local_'+ str(L) + '_' + str(T) + '.txt'
    if os.path.isfile(filename):
        f = open(filename, 'r')
        S = []
        for line in f:
            S.append(int(line))
        f.close()
        print( 'starting from file', filename)
    else:
        S = [random.choice([1, -1]) for k in range(N)]
        print( 'starting from scratch')
    p  = 1.0 - math.exp(-2.0 / T)
    nsteps = 50000
    S = [random.choice([1, -1]) for k in range(N)]
    E = [energy(S, N, nbr)]
    for step in range(nsteps):
        k = random.randint(0, N - 1)
        Pocket, Cluster = [k], [k]
        while Pocket != []:
            j = Pocket.pop()
            for l in nbr[j]:
                if S[l] == S[j] and l not in Cluster \
                       and random.uniform(0.0, 1.0) < p:
                    Pocket.append(l)
                    Cluster.append(l)
        for j in Cluster:
            S[j] *= -1
        E.append(energy(S, N, nbr))
        if step%(nsteps//3) == 0:
            print( "L = %d \t %.8f%%"%(L,step/(nsteps/100.0)))

    E_mean = sum(E)/ len(E)
    E2_mean = sum(a ** 2 for a in E) / len(E)
    cv = (E2_mean - E_mean ** 2 ) / N / T ** 2
    CV += [cv]
    
    f = open(filename, 'w')
    for a in S:
       f.write(str(a) + '\n')
    f.close()
```
